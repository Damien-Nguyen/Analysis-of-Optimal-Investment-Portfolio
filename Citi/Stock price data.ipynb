```python
import os
from pathlib import Path
import requests
import pandas as pd
from dotenv import load_dotenv
import alpaca_trade_api as tradeapi

%matplotlib inline
```

```python
load_dotenv()
```

```python
alpaca_api_key=os.getenv("ALPACA_API_KEY")
alpaca_secret_key=os.getenv("ALPACA_SECRET_KEY")
alpaca=tradeapi.REST(
    alpaca_api_key,
    alpaca_secret_key,
    api_version="v2")
```

```python
start_d = pd.Timestamp("2022-01-01", tz="America/New_York").isoformat()
end_d = pd.Timestamp("2023-03-31", tz="America/New_York").isoformat()

tickers = ["C"]

timeframe = "1Month"

df= alpaca.get_bars(
    tickers,
    timeframe,
    start = start_d,
    end = end_d
).df


df = df.drop(columns=["open","high","low","volume","trade_count","vwap"])

df = df[df['symbol']=='C'].drop('symbol', axis=1)

# check for NaN values
df.dropna()
df.rename(columns = {'close':'Citi'}, inplace = True)

# Preview DataFrame
df
```

```python
df["Citigroup Actual Returns"] = df["Citi"].pct_change()
df.head()
```

```python
short_window = 2
long_window = 4
df["Citi SMA_Fast"] = df["Citi"].rolling(window=short_window).mean()

df["Citi SMA_Slow"] = df["Citi"].rolling(window=long_window).mean()

df.dropna(inplace=True)

Abigail_df = df
Abigail_df.head()
```
